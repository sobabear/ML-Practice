# sigmoid

```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# example 
x = np.array([1.0, 0.5, -1, 2])
sigmoid(x)
```

```python
import matplotlib.pyplot as plt

test = np.arange(-5, 5, 0.1)
result = sigmoid(test)

plt.plot(test, result)
plt.ylim(-0.1, 1.1)
plt.show()

```

```python
def step_function(x):
  return np.array(x > 0, dtype=int)


x1 = np.arange(-5, 5, 0.1)
y1 = step_function(x1)
y2 = sigmoid(x1)

plt.plot(x1, y2, label='sigmoid')
plt.plot(x1, y1, linestyle='--', label='step')
plt.xlabel('x')
plt.ylabel('y')

plt.title('sigmoid & step')
plt.legend()
plt.show()
```

신경망에서는 activiation 함수로 non-linear 함수를 사용해야한다. 선형함수를 사용하면 층을 깊게 하는 의미가 없어지기 때문.
